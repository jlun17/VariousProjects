```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
Year = np.arange(1950, 2018)
T50_60 = np.array([114, 114, 114, 114, 114, 117, 117, 126, 126, 126])
T60_70 = np.array([141, 141, 141, 141, 225, 225, 225, 225, 270, 270])
T70_80 = np.array([270, 395, 396, 430, 435, 476, 521, 549, 565, 605])
T80_90 = np.array([619, 742, 860.72, 1008, 1159, 1275, 1390, 1474, 1544, 1636])
T90_00 = np.array([1734, 1855, 1982, 2022, 2124, 2174, 2224, 2258, 2286, 2340])
T00_10 = np.array([2408, 2502, 2655, 2907.90, 2939.76, 3381.3, 3465.9, 4040, 4424, 4822])
T10_18 = np.array([5256, 6307.2, 6874.8, 7493.6, 7868.2, 8301, 8716, 9152])
Tuition = np.array([])
Tuition = np.append(Tuition, T50_60)
Tuition = np.append(Tuition, T60_70)
Tuition = np.append(Tuition, T70_80)
Tuition = np.append(Tuition, T80_90)
Tuition = np.append(Tuition, T90_00)
Tuition = np.append(Tuition, T00_10)
Tuition = np.append(Tuition, T10_18)

fig = plt.figure(figsize=(10,6))
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(Year, Tuition, label=r"$Tuition$", color="green")
axes.legend(loc=2,fontsize=14)
axes.set_xlabel(r"$Year$",fontsize=18)
axes.set_ylabel(r"$Tuirion$",fontsize=18)
axes.set_title('CSU Tuition Rates no infaltion')
```

```python
C50_60 = np.array([24.1, 26, 26.5, 26.7, 26.9, 26.8, 27.2, 28.1, 28.9, 29.1])
C60_70 = np.array([29.6, 29.9, 30.2, 30.6, 31.0, 31.5, 32.4, 33.4, 34.8, 36.7])
C70_80 = np.array([38.8, 40.5, 41.8, 44.4, 49.3, 53.8, 56.9, 60.6, 65.2, 72.6])
C80_90 = np.array([82.4, 90.9, 96.5, 99.6, 103.9, 107.6, 109.6, 113.6, 118.3, 124])
C90_00 = np.array([130.7, 136.2, 140.3, 144.5, 148.2, 152.4, 156.9, 160.5, 163, 166.6])
C00_10 = np.array([172.2, 177.1, 179.9, 184, 188.9, 195.3, 201.6, 207.3, 215.303, 214.537])
C10_18 = np.array([218.056, 224.939, 229.594, 232.957, 236.736, 237.017, 240.007, 245.12])
CPI = np.array([])
CPI = np.append(CPI, C50_60)
CPI = np.append(CPI, C60_70)
CPI = np.append(CPI, C70_80)
CPI = np.append(CPI, C80_90)
CPI = np.append(CPI, C90_00)
CPI = np.append(CPI, C00_10)
CPI = np.append(CPI, C10_18)
CPI19 = 255.657
AdjustedTuition = Tuition * (CPI19 / CPI)

fig = plt.figure(figsize=(10,6))
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(Year, Tuition, label=r"$Tuition$", color="green")
axes.plot(Year, AdjustedTuition, label=r"$AdjTuition$", color="blue")
axes.plot(Year, CPI, label=r"$inflation$", color="red")
axes.legend(loc=2,fontsize=14)
axes.set_xlabel(r"$Year$",fontsize=18)
axes.set_ylabel(r"$Tuirion$",fontsize=18)
axes.set_title('CSU Tuition Rates')

fig = plt.figure(figsize=(10,6))
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(Year, Tuition, label=r"$Tuition$", color="green")
axes.legend(loc=2,fontsize=14)
axes.set_xlabel(r"$Year$",fontsize=18)
axes.set_ylabel(r"$AdjTuition$",fontsize=18)
axes.set_title('CSU Tuition Rates with infaltion')

fig = plt.figure(figsize=(10,6))
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(Year, AdjustedTuition, label=r"$AdjTuition$", color="blue")
axes.legend(loc=2,fontsize=14)
axes.set_xlabel(r"$Year$",fontsize=18)
axes.set_ylabel(r"$AdjTuition$",fontsize=18)
axes.set_title('CSU Tuition Rates with infaltion')

fig = plt.figure(figsize=(10,6))
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(Year, CPI, label=r"$inflation$", color="red")
axes.legend(loc=2,fontsize=14)
axes.set_xlabel(r"$Year$",fontsize=18)
axes.set_ylabel(r"$inflation$",fontsize=18)
axes.set_title('inflation')
```
